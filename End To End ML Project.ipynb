```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
dataset = pd.DataFrame(data,columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTART'])
```

```python
dataset['Price'] = target
```

```python
dataset.info()
```

```python
dataset.describe()
```

check missing value

```python
dataset.isnull().sum()
```

correlation

```python
dataset.corr()
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
sns.regplot(x='TAX',y='RM',data=dataset)
```

Devide independant feature and dependant feature

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
##train test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

standardize the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

apply linear regression

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

predict

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
```

prediction with residual

```python
residuals = y_test-y_pred
sns.displot(residuals,kind="kde")
```

```python
plt.scatter(y_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
np.sqrt(mean_absolute_error(y_test,y_pred))
```

R square and adjusted r square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```

```python
ad_score = 1-((1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1))
ad_score
```

new data predict

```python
new_data = data[0].reshape(1,-1)
new_data = scaler.transform(new_data)
regression.predict(new_data)
```
